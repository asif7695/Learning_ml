```python
import pandas as pd
df = pd.read_csv("/content/tips-dataset/tip.csv")
df.head()
```

```python
import seaborn as sns
```

```python
#1. scatterplot (numarical-numarical)
sns.scatterplot(data=df,x=df["total_bill"],y=df["tip"],hue="sex",style=df["smoker"],size=df["size"])
```

```python
#2. barplot (numarical-catagorical)
titanic = pd.read_csv("/content/Titanic-Dataset.csv")
titanic.head()
```

```python
sns.barplot(data=titanic, x=titanic["Pclass"], y=titanic["Age"],hue=titanic["Sex"])
```

```python
#3. boxplot(num-cat)
sns.boxplot(data=titanic, x=titanic["Sex"],y=titanic["Age"],hue=titanic["Survived"])
```

```python
#4 distplot
sns.distplot(titanic[titanic["Survived"]==0]["Age"])
sns.distplot(titanic[titanic["Survived"]==1]["Age"])
```

```python
#5. heatmap (cat-cat)
sns.heatmap(pd.crosstab(titanic["Sex"],titanic["Survived"]))
```

```python
#6. clustermap
sns.clustermap(pd.crosstab(titanic["SibSp"],titanic["Survived"]))
```

```python
#7. pairplot
sns.pairplot(titanic)
```

```python
#8. lineplot
sns.lineplot(data=df,x=df["total_bill"],y=df["tip"])
```
