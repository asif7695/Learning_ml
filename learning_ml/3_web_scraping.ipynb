```python
final = pd.DataFrame()
for j in range(1,10):
  webpage = requests.get(f"https://remote.co/remote-jobs/accounting?page={j}").text
  soup = BeautifulSoup(webpage, 'lxml')
  jobs = soup.find_all('div',class_="sc-14nyru2-2 fmkHkh")
  job_title = []
  location = []
  time = []
  salary = []

  for i in jobs:
    job_title.append(i.find('a').text)
    location.append(i.find('span',class_="sc-jv5lm6-10 efTTiy allowed-location").text)
    time.append(i.find('li',class_="sc-jv5lm6-9 jxMgfz tag-name").text)
  d = {'job_title':job_title,'location':location,'tags':time}
  df = pd.DataFrame(d)
  final = pd.concat([final,df])
```
