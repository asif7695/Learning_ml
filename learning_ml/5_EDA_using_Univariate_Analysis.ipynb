```python
import pandas as pd

df = pd.read_csv("/content/Titanic-Dataset.csv")
df.head()
```

```python
df = df.drop(["PassengerId"], axis=1)
df.head()
```

```python
# 1. catagorical data
# a. countplot
import seaborn as sns
#sns.countplot(df['Sex'])
df["Pclass"].value_counts().plot(kind='bar')
```
