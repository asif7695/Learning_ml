```python
main_df = pd.DataFrame()
for i in range(1,2113):
  response = requests.get(f"https://kitsu.io/api/edge/anime?page%5Bnumber%5D={i}&page%5Bsize%5D=10")
  temp_data = pd.json_normalize(response.json()['data'])[['id','attributes.canonicalTitle','attributes.status','attributes.subtype','attributes.averageRating','attributes.nsfw','attributes.popularityRank','attributes.ratingRank','attributes.ageRatingGuide','attributes.episodeCount','attributes.userCount','attributes.favoritesCount','attributes.startDate']]
  main_df = pd.concat([main_df,temp_data], ignore_index=True)
```
